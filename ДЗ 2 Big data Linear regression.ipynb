```python
import pandas as pd
```

```python
df = pd.read_csv('homework.csv')
```

```python
"""
This data frame contains the following columns:

crim - per capita crime rate by town.
zn - proportion of residential land zoned for lots over 25,000 sq.ft.
indus - proportion of non-retail business acres per town.
chas - Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).
nox - nitrogen oxides concentration (parts per 10 million).
rm - average number of rooms per dwelling.
age - proportion of owner-occupied units built prior to 1940.
dis - weighted mean of distances to five Boston employment centres.
rad - index of accessibility to radial highways.
tax - full-value property-tax rate per \$10,000.
ptratio - pupil-teacher ratio by town.
black - 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.
lstat - lower status of the population (percent).
medv - median value of owner-occupied homes in \$1000s.
"""
df.head()
```

```python
df.info()
# заполнять пропущенные значения не нужно. Все значения числовые, преобразовывать не нужно
```

```python
df.columns
```

```python
X_columns = list(df.columns)[1:14]
```

```python
# Формируем признаки и целевую переменную
X, y = df[X_columns], df['medv']
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train = pd.DataFrame(X_train, columns=X.columns)
X_test = pd.DataFrame(X_test, columns=X.columns)
X_train.head()
```

#### 1. LinearRegression

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
y_predict_lr = linreg.predict(X_test)
results = pd.DataFrame({'target': y_test, 'predict_lr': y_predict_lr}).reset_index()[['target','predict_lr']]
results.head()
```

Оценим качество

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
print('mean_squared_error - ', mean_squared_error(y_test, linreg.predict(X_test)),'\n',
      'r2_score - ', r2_score(y_test, linreg.predict(X_test)),'\n',
      'mean_absolute_error - ', mean_absolute_error(y_test, linreg.predict(X_test)))
```

#### 2. SVR

```python
from sklearn.svm import SVR
svrmodel = SVR()
svrmodel.fit(X_train, y_train)
```

```python
y_predict_svr = svrmodel.predict(X_test)
results['predict_svr'] = y_predict_svr
results
```

Оценим качество

```python
print('mean_squared_error - ', mean_squared_error(y_test, svrmodel.predict(X_test)),'\n',
      'r2_score - ', r2_score(y_test, svrmodel.predict(X_test)),'\n',
      'mean_absolute_error - ', mean_absolute_error(y_test, svrmodel.predict(X_test)))
```

#### 3. DecisionTreeRegressor

```python

from sklearn.tree import DecisionTreeRegressor
treereg = DecisionTreeRegressor(random_state=0)
treereg.fit(X_train, y_train)
```

```python
y_predict_treereg = treereg.predict(X_test)
results['predict_treereg'] = y_predict_treereg
results
```

Оценим качество

```python
print('mean_squared_error - (0 - best)', mean_squared_error(y_test, treereg.predict(X_test)),'\n',
      'r2_score - (1 - best)', r2_score(y_test, treereg.predict(X_test)),'\n',
      'mean_absolute_error - (0 - best)', mean_absolute_error(y_test, treereg.predict(X_test)))
```

#### Сводная таблица оценок

```python
quality=pd.DataFrame({'mean_squared_error':[mean_squared_error(y_test, linreg.predict(X_test)), mean_squared_error(y_test, svrmodel.predict(X_test)), mean_squared_error(y_test, treereg.predict(X_test))],
                      'r2_score':[r2_score(y_test, linreg.predict(X_test)),r2_score(y_test, svrmodel.predict(X_test)),r2_score(y_test, treereg.predict(X_test))],
                      'mean_absolute_error':[mean_absolute_error(y_test, linreg.predict(X_test)),mean_absolute_error(y_test, svrmodel.predict(X_test)),mean_absolute_error(y_test, treereg.predict(X_test))]}, 
                     index = ['linreg', 'svr', 'treereg'])

#лучший результат у DecisionTreeRegressor, худший у svr
quality
```
